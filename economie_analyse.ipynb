```python
# analise exploratoria das variaveis para poder excluir outliers antes da normalização
variaveis = ['pib_per_capita', 'chomage', 'inflation', 'dette_publique',
       'croissance', 'balance_commerciale', 'salaire_median', 'pauvrete',
       'gini', 'depenses_logement', 'privation_materielle',
       'IDH']

# boxplot (visual)
# criação da estrutura gráfica principal (imagem)
# figsize=(16, 12) define a largura e a altura da imagem final em polegadas.
# nrows=3 e ncols=4 criam a matriz de 12 espaços.
fig, axes = plt.subplots(nrows=3, ncols=4, figsize=(16, 12))

# o comando .flatten() converte essa matriz em uma lista linear de 12 posições (0 a 11).
# isso é necessário para automatizar a plotagem usando um loop simples.
axes = axes.flatten()

# loop de repetição para plotagem
# O 'for' inicia a repetição. 
# A variável 'i' recebe o número da posição (0, 1, 2...).
# A variável 'var' recebe o nome da coluna ('pib_per_capita', 'chomage'...).
# o enumerate() serve para extrair duas informações simultaneamente da sua lista: a posição numérica e o conteúdo de texto.
# i e var não fazem nada sozinhas; elas são apenas nomes genéricos (caixas vazias) que você criou para guardar o que o loop vai cuspir. 
# # O verdadeiro motor que gera a posição e o texto é o enumerate().
for i, var in enumerate(variaveis):
    
    # --- INÍCIO DA PRIMEIRA VOLTA DO LOOP ---
    # O Python substitui as variáveis pelos valores da vez:
    # i = 0
    # var = 'pib_per_capita'
    
    # x=df[var] se transforma em x=df['pib_per_capita']. O gráfico recebe apenas os dados desta coluna.
    # ax=axes[i] se transforma em ax=axes[0]. Isso avisa ao Python para colocar o desenho exatamente no primeiro espaço livre da sua grade de 12 imagens.
    sns.boxplot(x=df[var], ax=axes[i])
    
    # axes[i].set_title(var) se transforma em axes[0].set_title('pib_per_capita'). 
    # Isso escreve 'pib_per_capita' no topo do primeiro quadrado.
    axes[i].set_title(var)
    
    # axes[i].set_xlabel('') se transforma em axes[0].set_xlabel('').
    # O Seaborn, por padrão, tenta escrever o nome da coluna no eixo horizontal. Esse comando substitui o nome por um texto vazio ('') para não poluir a tela.
    axes[i].set_xlabel('')

    # --- FIM DA PRIMEIRA VOLTA ---
    
    # O loop retorna para o topo (para a linha do 'for') e pega o próximo item da lista.
    # Na segunda volta:
    # i = 1
    # var = 'chomage'
    # O processo se repete de forma idêntica até que as 12 variáveis da lista tenham sido processadas.
```

```python
# IQR
# Inicia o loop: a cada volta, 'var' assume o nome de uma coluna da lista
for var in variaveis:
    
    # 1. Calcula os quartis especificamente para a coluna da iteração atual
    Q1 = df[var].quantile(0.25)
    Q3 = df[var].quantile(0.75)
    
    # 2. Calcula o IQR
    IQR = Q3 - Q1
    
    # 3. Define os limites matemáticos de normalidade
    limite_inferior = Q1 - (1.5 * IQR)
    limite_superior = Q3 + (1.5 * IQR)
    
    # 4. Filtra a base original buscando os países fora dos limites
    outliers = df[(df[var] < limite_inferior) | (df[var] > limite_superior)]
    
    # 5. Exibição dos resultados estruturada
    print(f"--- Outliers na variável: {var} ---")
    
    # Condição lógica para evitar poluição visual no terminal
    if not outliers.empty:
        # Se houver outliers, imprime o nome do país e o valor anômalo
        print(outliers[[var]])
    else:
        # Se o filtro retornar vazio, informa que a distribuição está normal
        print("Nenhum outlier identificado.")
        
    # Pula uma linha no terminal antes de iniciar a próxima variável
    print("\n")
```

```python
# tratamento de outliers
# Ao travar o valor bruto no limite imposto pelo IQR, você transforma o outlier no novo valor máximo aceitável (Xmax) daquela variável. 
# # Como a fórmula penaliza quem atinge o Xmax, a nota padronizada desse país cai para 0. (caso para valores que quanto mais alto pior)

# Captura automaticamente as 12 colunas numéricas do seu dataframe
variaveis = df.columns.tolist()

# Inicia a repetição para tratar cada coluna isoladamente
for var in variaveis:
    
    # 1. Calcula os quartis e o IQR da coluna atual
    Q1 = df[var].quantile(0.25)
    Q3 = df[var].quantile(0.75)
    IQR = Q3 - Q1
    
    # 2. Define as fronteiras matemáticas
    limite_inferior = Q1 - (1.5 * IQR)
    limite_superior = Q3 + (1.5 * IQR)
    
    # 3. Aplica o travamento (Capping)
    # O comando abaixo pega a coluna original, corta os excessos e a sobrescreve 
    # já com os dados tratados.
    df[var] = df[var].clip(lower=limite_inferior, upper=limite_superior)

# Confirmação visual no terminal
print("Tratamento de outliers concluído com sucesso para todas as variáveis.")
```

```python
# normalização
# 1. Separação manual obrigatória pela direção do impacto
var_positivas = [
    'pib_per_capita', 'salaire_median', 'croissance', 
    'balance_commerciale', 'IDH'
]

var_negativas = [
    'chomage', 'inflation', 'dette_publique', 
    'pauvrete', 'gini', 'depenses_logement', 'privation_materielle'
]

# 2. Loop de Normalização para Variáveis Positivas (Maior é Melhor)
for var in var_positivas:
    minimo = df[var].min()
    maximo = df[var].max()
    
    # Aplica a fórmula padrão e cria uma nova coluna
    # Ex: a coluna 'idh' vai gerar uma nova coluna chamada 'idh_norm'
    df[f'{var}_norm'] = (df[var] - minimo) / (maximo - minimo)

# 3. Loop de Normalização para Variáveis Negativas (Menor é Melhor)
for var in var_negativas:
    minimo = df[var].min()
    maximo = df[var].max()
    
    # Aplica a fórmula invertida e cria uma nova coluna
    df[f'{var}_norm'] = (maximo - df[var]) / (maximo - minimo)

print("Normalização concluída. Todas as colunas '_norm' estão na escala de 0 a 1.")
```

```python
# criando o ranking - les classements
# 1. Cálculo do Score de Richesse
# Calcula a média da linha (axis=1) apenas para as colunas listadas
df['score_richesse'] = df[['pib_per_capita_norm', 'salaire_median_norm']].mean(axis=1)

# 2. Cálculo do Score de Stabilité Économique
df['score_stabilite'] = df[[
    'croissance_norm', 'balance_commerciale_norm', 
    'inflation_norm', 'chomage_norm', 'dette_publique_norm'
]].mean(axis=1)

# 3. Cálculo do Score de Qualité de Vie
# Atenção: Utilizamos 'IDH_norm' com letras maiúsculas conforme você corrigiu
df['score_qualite_vie'] = df[[
    'IDH_norm', 'gini_norm', 'pauvrete_norm', 
    'depenses_logement_norm', 'privation_materielle_norm'
]].mean(axis=1)

# --- Exibição dos Rankings ---

# Cria uma tabela ordenada do maior para o menor (ascending=False)
ranking_richesse = df.sort_values(by='score_richesse', ascending=False)
print("--- TOP 5: CLASSEMENT RICHESSE ---")
print(ranking_richesse[['score_richesse']].head(5))
print("\n")

ranking_stabilite = df.sort_values(by='score_stabilite', ascending=False)
print("--- TOP 5: CLASSEMENT STABILITÉ ÉCONOMIQUE ---")
print(ranking_stabilite[['score_stabilite']].head(5))
print("\n")

ranking_qualite = df.sort_values(by='score_qualite_vie', ascending=False)
print("--- TOP 5: CLASSEMENT QUALITÉ DE VIE ---")
print(ranking_qualite[['score_qualite_vie']].head(5))
```

Na estatística macroeconômica, quando você agrupa variáveis normalizadas dessa forma, o resultado final passa a ser chamado de Índice Composto.

A principal implicação conceitual de utilizar uma média simples é que o modelo assume que todas as variáveis possuem exatamente a mesma importância (o mesmo peso) dentro da sua respectiva categoria.

```python
# correlation
# Calcula a correlação de Pearson (padrão)
matriz_correlacao = df[variaveis].corr()

# Exibe o resultado numérico
print(matriz_correlacao)
```

```python
# Define a dimensão da imagem gerada. Como são 12 variáveis, um tamanho de 12x10 evita que os textos fiquem sobrepostos.
plt.figure(figsize=(12, 10))

# Gera o mapa de calor (Heatmap) com as configurações visuais:
# matriz_correlacao: fornece os dados numéricos calculados acima.
# annot=True: força o gráfico a desenhar o valor numérico dentro de cada célula.
# cmap='coolwarm': utiliza azul para correlação negativa, branco para zero, e vermelho para positiva.
# fmt=".2f": formata os números para exibirem apenas duas casas decimais.
# vmin=-1, vmax=1: fixa a legenda lateral estritamente nos limites matemáticos da correlação.
sns.heatmap(matriz_correlacao, annot=True, cmap='coolwarm', fmt=".2f", vmin=-1, vmax=1)

# Renderiza e exibe o gráfico finalizado na tela do VS Code.
plt.show()
```

```python
# separando correlações fortes

# Define o limite estatístico para considerar a correlação forte
limite = 0.7

# 1. Remove o espelhamento e a diagonal principal
# np.triu (triangle upper) com k=1 gera uma máscara que isola apenas a metade superior da matriz. 
# Isso evita que o resultado mostre o mesmo par de variáveis duas vezes.
mascara = np.triu(np.ones(matriz_correlacao.shape), k=1).astype(bool)
matriz_filtrada = matriz_correlacao.where(mascara)

# 2. Transforma a grade 2D em uma lista estruturada
# O método .stack() pega os valores que sobraram na matriz e os empilha em um formato de colunas. 
# O .reset_index() converte os nomes das variáveis em colunas reais.
pares = matriz_filtrada.stack().reset_index()
pares.columns = ['Variavel_1', 'Variavel_2', 'Correlacao']

# 3. Aplica o filtro matemático
# O método .abs() (valor absoluto) é crucial aqui. Ele trata números negativos 
# como positivos temporariamente durante a avaliação. Assim, uma correlação de -0.85 
# será identificada como forte e passará pelo filtro do limite de 0.7.
correlacoes_fortes = pares[pares['Correlacao'].abs() >= limite]

# Ordena os resultados e descarta a numeração antiga da tabela original
correlacoes_fortes = correlacoes_fortes.sort_values(by='Correlacao', key=abs, ascending=False).reset_index(drop=True)

print("--- Pares com Correlação Forte (>= 0.7 ou <= -0.7) ---")
print(correlacoes_fortes)
```

```python
# Fase 1: O Método do Cotovelo (Elbow Method)
from sklearn.cluster import KMeans

# 1. Isola apenas as colunas normalizadas para o cálculo
# Esta linha cria uma lista automática pegando qualquer coluna que termine com '_norm'
colunas_para_cluster = [col for col in df.columns if col.endswith('_norm')]
dados_cluster = df[colunas_para_cluster]

# 2. Cria uma lista vazia para guardar o erro (inércia) de cada teste
inercia = []

# 3. Inicia um loop testando a divisão da Europa em 1 até 10 grupos
for k in range(1, 11):
    # Configura o algoritmo para o 'k' atual da rodada. 
    # random_state=42 garante que o resultado seja o mesmo toda vez que você rodar.
    kmeans_teste = KMeans(n_clusters=k, random_state=42, n_init=10)
    
    # Executa a matemática nos dados normalizados
    kmeans_teste.fit(dados_cluster)
    
    # Salva o resultado da inércia na lista
    inercia.append(kmeans_teste.inertia_)

# 4. Plota o gráfico de linha com os resultados
plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), inercia, marker='o')
plt.title('Elbow Method pour définir la quantité de clusters')
plt.xlabel('Número de Grupos (k)')
plt.ylabel('Inércia (Distância Interna)')
plt.grid(True)
plt.show()
```

Analisando os dados friamente, sem nenhum contexto indica que não é indicado agrupar esses países. Dado que viola a Teoria das Zonas Monetárias Otimais (ZMO), desenvolvida pelo Prêmio Nobel Robert Mundell (1961). Mundell postula que, para compartilhar uma moeda de forma eficiente, as economias precisam reagir de forma semelhante aos choques externos. Os dados extraídos do modelo K-Means comprovam que a Europa está dividida em blocos de desempenho diferentes.

Porém, o bloco funciona por conta de:

1. Interdependência Estrutural
A pertinência real do bloco fundamenta-se não na homogeneidade, mas na dependência mútua. A literatura econômica reconhece uma dinâmica de compensação cruzada: as economias de alto desempenho necessitam da inclusão de economias mais frágeis para evitar uma valorização excessiva do Euro, garantindo assim a competitividade global de suas exportações. Em contrapartida, as economias instáveis abdicam da soberania monetária em troca da credibilidade institucional do BCE, blindando-se contra crises cambiais e captando crédito internacional a custos significativamente inferiores.

2. A Teoria Endógena e os Mecanismos de Sustentação
Para neutralizar as assimetrias comprovadas nos dados, a Zona do Euro opera sob o conceito de ZMO Endógena, teorizado pelos economistas Jeffrey Frankel e Andrew Rose (1998). Esta vertente defende que o compartilhamento da moeda força a integração comercial e institucional ao longo do tempo. O bloco compensa suas diferenças estruturais através de critérios estabelecidos por outros teóricos da ZMO:

- Mobilidade de Fatores (Critério de Mundell): O Espaço Schengen funciona como uma válvula de escape essencial. Diante da impossibilidade de desvalorizar a moeda em crises locais, a força de trabalho migra das zonas estagnadas para as zonas em expansão, equilibrando o mercado.

- Transferências Fiscais (Critério de Peter Kenen, 1969): Kenen afirma que uniões monetárias assimétricas sobrevivem através de um orçamento centralizado para redistribuição de renda. A União Europeia simula este mecanismo através dos Fundos de Coesão e do Mecanismo Europeu de Estabilidade (MEE), transferindo capital das economias líderes para o desenvolvimento estrutural dos demais blocos.

Conclui-se que a pertinência atual do Euro não deriva da semelhança natural de seus membros, mas da pesada infraestrutura de integração concebida para atenuar as disparidades, tornando os custos de uma eventual dissolução muito superiores aos ônus da atual assimetria.

Une analyse purement objective des données, sans éléments de contexte, indique qu'il n'est pas recommandé de regrouper ces pays. Mundell postule que pour partager efficacement une monnaie, les économies doivent réagir de manière similaire aux chocs externes. Les données extraites du modèle K-Means prouvent que l'Europe est divisée en différents blocs de performance.

Cependant, le bloc fonctionne grâce aux éléments suivants :

1. Interdépendance structurelle
La pertinence réelle du bloc repose non pas sur l'homogénéité, mais sur la dépendance mutuelle. La littérature économique reconnaît une dynamique de compensation croisée : les économies très performantes ont besoin d'inclure des économies plus fragiles afin d'éviter une appréciation excessive de l'euro, garantissant ainsi la compétitivité mondiale de leurs exportations. En contrepartie, les économies instables renoncent à leur souveraineté monétaire en échange de la crédibilité institutionnelle de la BCE, se protégeant ainsi des crises de change et obtenant des crédits internationaux à des coûts nettement inférieurs.

2. La théorie endogène et les mécanismes de soutien
Afin de neutraliser les asymétries prouvées par les données, la zone euro fonctionne selon le concept de ZMO endogène, théorisé par les économistes Jeffrey Frankel et Andrew Rose (1998). Cette approche soutient que le partage de la monnaie force l'intégration commerciale et institutionnelle au fil du temps. Le bloc compense ses différences structurelles grâce à des critères établis par d'autres théoriciens de la ZMO :

- Mobilité des facteurs (Critère de Mundell) : L'espace Schengen fonctionne comme une soupape de sécurité essentielle. Face à l'impossibilité de dévaluer la monnaie lors de crises locales, la main-d'œuvre migre des zones stagnantes vers les zones en expansion, équilibrant ainsi le marché.
- Transferts fiscaux (Critère de Peter Kenen, 1969) : Kenen affirme que les unions monétaires asymétriques survivent grâce à un budget centralisé pour la redistribution des revenus. L'Union européenne simule ce mécanisme par le biais des fonds de cohésion et du Mécanisme européen de stabilité (MES), en transférant des capitaux des économies leaders vers le développement structurel des autres blocs.

En conclusion, la pertinence actuelle de l'euro ne découle pas d'une similitude naturelle entre ses membres, mais plutôt de la lourde infrastructure d'intégration conçue pour atténuer les disparités, rendant les coûts d'une éventuelle dissolution bien supérieurs aux charges de l'asymétrie actuelle.
